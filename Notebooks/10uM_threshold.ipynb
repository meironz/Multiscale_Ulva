```python
# Import of I data
# irradiance data is based on meteorological data from: https://ims.data.gov.il/he/ims/6 
dFI = pd.read_csv('../data/ims_data_2014_umol_photons.csv',encoding= 'unicode_escape',header=None)
days = list(range(1,366))
```

```python
I_plot = np.zeros(24*365)
light_hours = list(range(5,19))
for day in days:
    for hour in light_hours:
        I_plot[(day - 1)*24 + hour] = float(dFI.iloc[day][hour-1])

annual_hours = list(range(1,(366-1)*24+1))
f0 = interpolate.interp1d(annual_hours, I_plot,kind = 'linear')
```

```python
# N and biomass Dynamics in 3 seasons - Winter n_reactor (different cultivation lengths - 14,9,11)

n_reactors = max(lengths)
NSEA, NEXT, NINT, M, Total_N, T = [],[],[],[],[],[]
Total_biomass = []
NSEA_seasons_f = []
Final_NSEA = []
NSEA_seasons, NINT_seasons, M_seasons = [],[],[]
Data = []
n_days_all = [14, 9, 11]
T_all = []

print('Starting point')


# setup initial conditions identical for all reactors
x0 = n_reactors*[Nsea, Next0, Nint0, m0]

dilution = 0
weeks = [1, 7, 45]
seasons = ['Winter', 'Spring', 'Autumn']
print(weeks)

for week0 in weeks:
    n_days = n_days_all[weeks.index(week0)]
    t0 = 10 + 7 * 24 * (week0-1)
    t = list(range(t0,t0 + int(n_days*24)))
    times = list(range(t0,t0 + int(n_days*24)))
    args = (Nintcrit,Nintmax,Nintmin,Vmax,Ks,KN,miu,S,Z,KI,K0,Ka,Topt,Tmin,Tmax,losses20,
            teta,Sopt,Smin,Smax, Qp, Qsea, Nsea,f1,f0,dilution,n,umol_to_percent_DW)

    # solve the ODEs using the new syntax
    sol = solve_ivp(multi_N_f_un, [t[0], t[-1]], x0, args=args, t_eval = t)

    # take the solution of the state variables:

    for i_reactor in range(n_reactors):
        NSEA, NEXT, NINT, M, N_farm, Total_N, T, N_farm, M_farm = [],[],[],[],[],[],[],[],[]
        for i in range(i_reactor+1):
            NSEA.append(sol.y[i*4,:]) #[umol N in sea / l]
            NEXT.append(sol.y[i*4+1,:]) #[umol N in reactor / l]
            NINT.append(sol.y[i*4+2,:]) #[% g N / g DW]
            M.append(sol.y[i*4+3,:]) #[g DW / l]
            M_farm.append((M[-1][-1] - m0) * 1.785) #[produced kg DW / reactor / 2 weeks]
            N_farm.append((M[-1][-1] * NINT[-1][-1] - m0 * Nint0) * 1.785 / 100) #[removed g N / reactor eq sea volume / 2 weeks]
        Total_biomass.append(round(sum(M_farm),1)) #[produced kg DW / farm / 2 weeks]
        Total_N.append(round(sum(N_farm),2)) #[removed g N / farm / 2 weeks]
        T.append(sol.t)
    Seasonal_production = sum(Total_biomass)
    Seasonal_sequestration = sum(Total_N) #[removed g N / farm / 2 weeks]
    Data.append([Seasonal_production,Seasonal_sequestration])    
    NSEA_seasons.append(NSEA[-1])
    NINT_seasons.append(NINT[-1])
    M_seasons.append(M[-1])
    T_all.append(T)
    x0 = n_reactors*[Nsea, Next0, Nint0, m0]

T1 = T_all[0]
T2 = T_all[1]
T3 = T_all[2]
plot_result_seasons_days(T1,T2,T3,NSEA_seasons,NINT_seasons,M_seasons,Nsea)
np.savetxt('Seasonal_total_values_140911.csv',Data,delimiter=',')         

print("Done")

n_days = 14
```
